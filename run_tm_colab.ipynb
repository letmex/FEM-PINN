# TM 热-力-相场训练（Colab 双模式）

这个 Notebook 支持两种模式：
- `quick20`: 快速验收（约 step=20）
- `strict100`: 中程严格验证（约 step=100）

执行顺序：
1. 环境检查
2.（可选）挂载 Drive
3. 获取代码
4. 安装依赖
5. 选择运行模式并自动 patch config（仅 Colab 临时副本）
6. 启动训练
7. 打包结果


```python
# 0) 环境检查
import os, sys, subprocess, pathlib, re, shutil, textwrap
print('Python:', sys.version)
try:
    import torch
    print('Torch:', torch.__version__)
    print('CUDA available:', torch.cuda.is_available())
    if torch.cuda.is_available():
        print('GPU:', torch.cuda.get_device_name(0))
except Exception as e:
    print('Torch not available yet:', e)
```

```python
# 1) 可选：挂载 Google Drive
USE_GOOGLE_DRIVE = False  # 需要 Drive 时改 True
if USE_GOOGLE_DRIVE:
    from google.colab import drive
    drive.mount('/content/drive')
    print('Drive mounted.')
else:
    print('Skip Drive mount.')
```

```python
# 2) 选择代码来源
# 可选：'git' / 'drive' / 'zip'
SOURCE_MODE = 'git'

REPO_URL = 'https://github.com/letmex/FEM-PINN.git'
REPO_BRANCH = 'main'

# drive 模式
DRIVE_REPO_PATH = '/content/drive/MyDrive/phase-field-fracture-with-pidl'

# zip 模式
ZIP_PATH = '/content/phase-field-fracture-with-pidl.zip'

TARGET_ROOT = pathlib.Path('/content')
REPO_NAME = REPO_URL.rstrip('/').split('/')[-1].replace('.git', '')
REPO_DIR = TARGET_ROOT / REPO_NAME

def run_cmd(cmd, check=True):
    print('Run:', ' '.join(cmd))
    if check:
        subprocess.check_call(cmd)
    else:
        return subprocess.call(cmd)

if REPO_DIR.exists():
    print(f'Repo already exists: {REPO_DIR}')
    if SOURCE_MODE == 'git':
        run_cmd(['git', '-C', str(REPO_DIR), 'fetch', 'origin'], check=False)
        run_cmd(['git', '-C', str(REPO_DIR), 'checkout', REPO_BRANCH], check=False)
        run_cmd(['git', '-C', str(REPO_DIR), 'pull', 'origin', REPO_BRANCH], check=False)
        run_cmd(['git', '-C', str(REPO_DIR), 'lfs', 'pull'], check=False)
else:
    if SOURCE_MODE == 'git':
        cmd = ['git', 'clone', '--depth', '1', '--branch', REPO_BRANCH, REPO_URL, str(REPO_DIR)]
        run_cmd(cmd)
        run_cmd(['git', '-C', str(REPO_DIR), 'lfs', 'pull'], check=False)
    elif SOURCE_MODE == 'drive':
        src = pathlib.Path(DRIVE_REPO_PATH)
        if not src.exists():
            raise FileNotFoundError(f'DRIVE_REPO_PATH not found: {src}')
        subprocess.check_call(['cp', '-r', str(src), str(REPO_DIR)])
    elif SOURCE_MODE == 'zip':
        zip_path = pathlib.Path(ZIP_PATH)
        if not zip_path.exists():
            raise FileNotFoundError(f'ZIP_PATH not found: {zip_path}')
        subprocess.check_call(['unzip', '-q', str(zip_path), '-d', str(TARGET_ROOT)])
    else:
        raise ValueError(f'Unsupported SOURCE_MODE: {SOURCE_MODE}')

print('Repo dir:', REPO_DIR)
print('README exists:', (REPO_DIR / 'README.md').exists())
```

```python
# 3) 安装依赖（兼容 py3.12 / Colab）
import sys, subprocess

def run(cmd, check=True):
    print('>>', ' '.join(cmd))
    if check:
        subprocess.check_call(cmd)
    else:
        return subprocess.call(cmd)

run([sys.executable, '-m', 'pip', 'install', '-U', 'pip', 'setuptools', 'wheel'])

# 优先安装与 Colab(py3.12) 兼容的基础包
run([sys.executable, '-m', 'pip', 'install',
     'numpy>=1.26,<3',
     'scipy>=1.11',
     'pandas>=2.1',
     'matplotlib>=3.8',
     'tqdm>=4.65',
     'gmshparser==0.2.0'])

# requirements.txt 若存在就尝试安装；失败不终止（避免旧钉版本在 py3.12 崩掉）
REQ = REPO_DIR / 'requirements.txt'
if REQ.exists():
    rc = run([sys.executable, '-m', 'pip', 'install', '-r', str(REQ)], check=False)
    if rc != 0:
        print('requirements.txt install failed, continue with compatible packages already installed.')
else:
    print('requirements.txt not found, skip.')

# torch: Colab 通常已预装，优先复用
try:
    import torch
    print('Torch already available:', torch.__version__, 'CUDA:', torch.cuda.is_available())
except Exception:
    run([sys.executable, '-m', 'pip', 'install', 'torch'])
    import torch
    print('Torch installed:', torch.__version__, 'CUDA:', torch.cuda.is_available())
```

```python
# 4) 进入示例目录并选择运行模式
EXAMPLE_DIR = REPO_DIR / 'examples' / 'TM_single_notch_tensile'
if not EXAMPLE_DIR.exists():
    raise FileNotFoundError(EXAMPLE_DIR)
os.chdir(EXAMPLE_DIR)
print('Working dir:', EXAMPLE_DIR)

# 运行模式：'quick20' / 'strict50' / 'strict100'
RUN_PROFILE = 'strict50'
print('RUN_PROFILE =', RUN_PROFILE)
```

```python
# 5) 按 profile 自动 patch config.py（仅当前 Colab 副本）
cfg = EXAMPLE_DIR / 'config.py'
bak = EXAMPLE_DIR / 'config.colab.bak.py'
if not bak.exists():
    shutil.copy2(cfg, bak)
    print('Backup created:', bak)
base = bak.read_text(encoding='utf-8')
txt = base

def rep(pattern, repl, s):
    n = re.subn(pattern, repl, s, flags=re.M)[1]
    if n == 0:
        raise RuntimeError(f'Pattern not found: {pattern}')
    return re.sub(pattern, repl, s, flags=re.M)

if RUN_PROFILE == 'quick20':
    txt = rep(r'^time_dict\s*=\s*\{.*\}\s*$', "time_dict = {'t_start': 0.0, 't_end': 0.20, 'dt': 0.01}", txt)
    txt = rep(r'^\s*"require_converged_step"\s*:\s*.*,$', "    \"require_converged_step\": False,", txt)
    txt = rep(r'^\s*"auto_physics_loop"\s*:\s*.*,$', "    \"auto_physics_loop\": False,", txt)
    txt = rep(r'^\s*"auto_physics_max_rounds"\s*:\s*.*,$', "    \"auto_physics_max_rounds\": 1,", txt)
elif RUN_PROFILE == 'strict50':
    txt = rep(r'^time_dict\s*=\s*\{.*\}\s*$', "time_dict = {'t_start': 0.0, 't_end': 0.50, 'dt': 0.01}", txt)
    txt = rep(r'^\s*"require_converged_step"\s*:\s*.*,$', "    \"require_converged_step\": True,", txt)
    txt = rep(r'^\s*"auto_physics_loop"\s*:\s*.*,$', "    \"auto_physics_loop\": True,", txt)
    txt = rep(r'^\s*"auto_physics_max_rounds"\s*:\s*.*,$', "    \"auto_physics_max_rounds\": 5,", txt)
elif RUN_PROFILE == 'strict100':
    txt = rep(r'^time_dict\s*=\s*\{.*\}\s*$', "time_dict = {'t_start': 0.0, 't_end': 1.00, 'dt': 0.01}", txt)
    txt = rep(r'^\s*"require_converged_step"\s*:\s*.*,$', "    \"require_converged_step\": True,", txt)
    txt = rep(r'^\s*"auto_physics_loop"\s*:\s*.*,$', "    \"auto_physics_loop\": True,", txt)
    txt = rep(r'^\s*"auto_physics_max_rounds"\s*:\s*.*,$', "    \"auto_physics_max_rounds\": 5,", txt)
else:
    raise ValueError('RUN_PROFILE must be quick20, strict50, or strict100')

cfg.write_text(txt, encoding='utf-8')
print('Patched config.py for profile:', RUN_PROFILE)
print('time_dict line:')
for line in txt.splitlines():
    if line.strip().startswith('time_dict ='):
        print('  ', line.strip())
        break
```

```python
# 6) 启动训练（自动准备 mesh + 可选后台运行）
import os, sys, shutil, subprocess
from pathlib import Path

EXAMPLE_DIR = REPO_DIR / 'examples' / 'TM_single_notch_tensile'
os.chdir(EXAMPLE_DIR)

# 自动准备 geo1_.msh 到 /content/geo1_.msh（当前 main.py 默认读取这里）
mesh_dst = Path('/content/geo1_.msh')
mesh_candidates = [
    REPO_DIR / 'geo1_.msh',
    EXAMPLE_DIR / 'geo1_.msh',
    Path('/content/geo1_.msh'),
    Path('/content/drive/MyDrive/geo1_.msh'),
]
mesh_src = None
for p in mesh_candidates:
    if p.exists():
        mesh_src = p
        break

if mesh_src is None:
    raise FileNotFoundError(
        '没找到 geo1_.msh。请把 geo1_.msh 上传到仓库根目录、/content，或 Google Drive。'
    )

if mesh_src.resolve() != mesh_dst.resolve():
    shutil.copy(mesh_src, mesh_dst)

print('mesh ready:', mesh_dst, 'size=', mesh_dst.stat().st_size)

RUN_ARGS = []  # 可选覆盖: ['6','120','1','AdaptiveReLU','1.0','-4.0']
BACKGROUND_RUN = False
MONITOR_LOG = False

cmd = [sys.executable, 'main.py'] + RUN_ARGS
print('Working dir:', EXAMPLE_DIR)
print('Run:', ' '.join(cmd))

if BACKGROUND_RUN:
    log_file = EXAMPLE_DIR / 'train.log'
    with open(log_file, 'w', encoding='utf-8') as f:
        proc = subprocess.Popen(cmd, stdout=f, stderr=subprocess.STDOUT)
    print('Started in background, PID=', proc.pid)
    print('Log file:', log_file)
    if MONITOR_LOG:
        print('--- tail -f train.log ---')
        subprocess.call(['tail', '-n', '80', '-f', str(log_file)])
else:
    proc = subprocess.run(cmd, text=True, capture_output=True)
    print('\n===== STDOUT (tail) =====\n')
    print(proc.stdout[-5000:] if proc.stdout else '(empty)')
    if proc.returncode != 0:
        print('\n===== STDERR (tail) =====\n')
        print(proc.stderr[-12000:] if proc.stderr else '(empty)')
        raise RuntimeError(f'main.py failed: returncode={proc.returncode}')
    print('\nTraining finished successfully.')
```

```python
# 7) 浏览最近结果（含中间进度）
import pandas as pd

items = sorted(EXAMPLE_DIR.iterdir(), key=lambda p: p.stat().st_mtime, reverse=True)
print('Recent files/folders:')
for p in items[:30]:
    print('-', p.name)

run_dirs = [p for p in items if p.is_dir() and p.name.startswith('hl_')]
if not run_dirs:
    print('\nNo hl_* run directory found yet.')
else:
    run = run_dirs[0]
    loss_csv = run / 'results' / 'losses' / 'loss_per_step.csv'
    trace_csv = run / 'results' / 'losses' / 'loss_trace.csv'
    phys_csv = run / 'results' / 'losses' / 'physics_consistency_per_step.csv'
    print('\nLatest run dir:', run)

    if loss_csv.exists():
        df = pd.read_csv(loss_csv)
        print('loss_per_step rows =', len(df))
        if len(df) > 0 and 'step' in df.columns:
            print('last completed step =', int(df['step'].max()))
        print('\nloss_per_step tail:')
        display(df.tail(5))
    else:
        print('loss_per_step.csv not found:', loss_csv)

    if phys_csv.exists():
        dph = pd.read_csv(phys_csv)
        print('\nphysics_consistency_per_step tail:')
        display(dph.tail(5))
    else:
        print('physics_consistency_per_step.csv not found:', phys_csv)

    if trace_csv.exists():
        dtr = pd.read_csv(trace_csv)
        print('\nloss_trace tail:')
        display(dtr.tail(10))
    else:
        print('loss_trace.csv not found:', trace_csv)
```

```python
# 8) 打包最近 run_* 目录（可选）
run_dirs = [p for p in EXAMPLE_DIR.iterdir() if p.is_dir() and p.name.startswith('run_')]
if run_dirs:
    latest_run = max(run_dirs, key=lambda p: p.stat().st_mtime)
    out_base = f'/content/{latest_run.name}'
    out_zip = shutil.make_archive(out_base, 'zip', root_dir=latest_run)
    print('Packed:', out_zip)
    if USE_GOOGLE_DRIVE:
        dst_dir = pathlib.Path('/content/drive/MyDrive/pinn_runs')
        dst_dir.mkdir(parents=True, exist_ok=True)
        dst = dst_dir / pathlib.Path(out_zip).name
        shutil.copy2(out_zip, dst)
        print('Copied to:', dst)
else:
    print('No run_* directory found.')
```

```python
# 9) 可选：恢复原始 config.py
RESTORE_CONFIG = False
if RESTORE_CONFIG:
    bak = EXAMPLE_DIR / 'config.colab.bak.py'
    cfg = EXAMPLE_DIR / 'config.py'
    if bak.exists():
        shutil.copy2(bak, cfg)
        print('config restored from backup')
    else:
        print('backup not found')
```
